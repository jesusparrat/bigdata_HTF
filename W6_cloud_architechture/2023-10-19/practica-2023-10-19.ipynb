```
# asignamos a file el archivo de texto que queremos trabajar y lo abrimos en una variable
file = 'dbfs:/FileStore/shared_uploads/jesuspt98@gmail.com/sensores.txt'
sensores = sc.textFile(file)
```

```
# observamos el archivo para ver con qué tratamos
sensores.collect()
```

Vemos que estamos tratando con strings por columnas, vamos a tratar esto:

```
# separamos los números por las comas
num_x_filas = sensores.flatMap(lambda elemento: elemento.split(',')) 
num_x_filas.collect()
```

```
# pasamos los números a flotantes desde string
num_x_filas_float = num_x_filas.map(float)
num_x_filas_float.take(15)
```

```
# hacemos una comprobación de los números testigos y vemos cuáles son para eliminarlos
numeros_neg = num_x_filas_float.filter(lambda elemento: elemento<0)
numeros_neg.take(50)
```

```
# filtramos y nos quedamos con los números entre 0 y 100
numeros = num_x_filas_float.filter(lambda elemento: 0<elemento<100)
numeros.take(40)
```

Ahora ya tenemos los datos filtrados como números flotantes guardados en una lista. Ahora podemos tratar los datos como queramos. Haremos la media con el comando mean() y también la haremos de manera manual, con la suma total de los elementos dividido entre el número de eventos. 

```
# hallamos la media de todos los números de la lista filtrada
media = numeros.mean()
# contamos los números totales que hay
numero_medidas = numeros.count()
# hacemos la suma de todos los números
suma_total = numeros.reduce(lambda elem1, elem2: elem1 + elem2)
# imprimimos por pantalla lo que necesitamos
print(f"La media es, por el parámetro media, {round(media,2)} y por la medida manual, {round(suma_total/numero_medidas,2)}")
```
